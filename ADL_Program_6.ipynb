6. Write a Python program to perform gradient descent on a linear function and plot the loss reduction over iterations.

```python
# Gradient descent on a linear model
import numpy as np
import matplotlib.pyplot as plt
```

```python
# synthetic data
rng = np.random.default_rng(1)
x = np.linspace(0, 10, 100)
true_w, true_b = 2.5, -1.0
y = true_w * x + true_b + rng.normal(0, 1, size=x.shape)
```

```python
# initialize params
w, b = 0.0, 0.0
lr = 0.01
epochs = 1000
losses = []
```

```python
for t in range(epochs):
    y_pred = w * x + b
    loss = np.mean((y_pred - y)**2)
    losses.append(loss)
    # gradients
    dw = 2 * np.mean((y_pred - y) * x)
    db = 2 * np.mean(y_pred - y)
    # update
    w -= lr * dw
    b -= lr * db

print(f"Learned parameters: w = {w:.3f}, b = {b:.3f}")
```

```python
# Plot loss
plt.figure(figsize=(6,4))
plt.plot(losses)
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.title("Loss Reduction over Iterations")
plt.grid(True)
plt.show()
```

```python
# Plot fitted line
plt.figure(figsize=(6,4))
plt.scatter(x, y, s=15, label='Data')
plt.plot(x, w*x + b, color='red', label='Fitted line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Fitted Line after Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```
