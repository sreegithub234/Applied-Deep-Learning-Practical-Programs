8. Apply PCA using Scikit-learn on the Iris dataset and visualize the reduced2D representation using a scatter plot.

**Scenario based question:**

"A botanist has measurements of iris flowers (sepal length, sepal width, petal length, petal width) for three species. To simplify visualization and analysis, we want to reduce the 4-dimensional data to 2D using PCA, then plot the flowers colored by species."

```python
# PCA on Iris dataset and 2D visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data  # shape (150, 4)
y = iris.target  # species labels: 0,1,2
target_names = iris.target_names
```

```python
# Standardize features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print("Explained variance ratio (2 components):", pca.explained_variance_ratio_)
```

```python
# Scatter plot
plt.figure(figsize=(6,6))
colors = ['r', 'g', 'b']
for target, color, name in zip([0,1,2], colors, target_names):
    plt.scatter(X_pca[y==target, 0], X_pca[y==target, 1],
                c=color, label=name, edgecolors='k', s=60)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of Iris Dataset (2D)")
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
