3. Implement a Python program to apply the sigmoid activation function on arrange of inputs and plot the output.

```python
# Sigmoid activation plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10, 400)
y = sigmoid(x)
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Activation Function")
plt.grid(True)
plt.show()
```
