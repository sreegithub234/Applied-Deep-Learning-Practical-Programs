7. Using NumPy, compute the eigenvalues and eigenvectors of a given square matrix and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  2x2 matrix  )
A = np.array([[2, 1],
              [1, 3]])
```

```python
# Compute eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
# Create a grid of points for visualization
x = np.linspace(-1.5, 1.5, 10)
y = np.linspace(-1.5, 1.5, 10)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])
```

```python
# Transform points using the matrix A
transformed_points = A @ points
```

```python
# Original points
plt.figure(figsize=(6,6))
plt.scatter(points[0,:], points[1,:], color='lightgray', label='Original points')
```

```python
# Plot eigenvectors (scaled for visibility)
origin = np.zeros((2,1))
for i in range(eigvecs.shape[1]):
    vec = eigvecs[:, i]
    plt.quiver(*origin, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=['r','b'][i], label=f'Eigenvector {i+1}')
```

```python
# Formatting
plt.xlim(-2, 2.5)
plt.ylim(-2, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Matrix Transformation & Eigenvectors")
plt.grid(True)
plt.legend()
plt.show()
```
